```python
import xml.etree.ElementTree as ET

def extract_features(data):
    root = ET.fromstring(data)
    document_id = root.attrib['id']
    publishing_date = root.attrib['publishingDate']
    title = root.find('./title').text
    lead_title = root.find('./leadTitle').text
    text = root.find('./text/page/text').text
    return {'document_id': document_id, 'publishing_date': publishing_date, 'title': title, 'lead_title': lead_title, 'text': text}

data = df_content_art.documentXML[0]
sample = extract_features(data)
```
